# Rate Equations

$$
\frac{dC_v}{dt}
= G - \{ \omega_i C_i + 2 \omega_i C_{2i} + 3 \omega_i C_{3i} + \omega_v C_s^v \} C_v
$$

$$
\frac{dC_i}{dt}
= G (1 - 2\varepsilon_{2i} - 3\varepsilon_{3i})
+ \omega_i C_{2i} (C_v + e_{2i})
+ \omega_i e_{3i} C_{3i}
- \{ C_v + C_i + 2C_{2i} + 3C_{3i} + C_s^i \} \omega_i C_i
$$

$$
\frac{dC_{2i}}{dt}
= \varepsilon_{2i} G
+ \tfrac{1}{2} \omega_i C_i^2
+ \omega_i C_v C_{3i}
- \omega_i C_{2i} (C_i + C_v + C_{2i} + C_{3i} + e_{2i} + C_s^i)
$$

$$
\frac{dC_{3i}}{dt}
= \varepsilon_{3i} G
+ 2 \omega_i C_i C_{2i}
- \omega_i C_{3i} (C_v + C_i + C_{2i} + C_{3i} + e_{3i} + C_s^i)
$$

$$
\frac{dC_{iL}^{111}}{dt}
= (1 - f_{100}(T))(\omega_i C_{3i} + 2 \omega_i C_{2i})
- k_{\mathrm{tr}}(T) C_{iL}^{111}
$$

$$
\frac{dC_{iL}^{100}}{dt}
= f_{100}(T)(\omega_i C_{3i} + 2 \omega_i C_{2i})
+ k_{\mathrm{tr}}(T) C_{iL}^{111}
$$

$$
\frac{dN_{\text{void}}}{dt}
= k_{\text{nuc}}(T) C_v^2 - \frac{N_{\text{void}}}{\tau_{\text{diss}}(T)}
$$

$$
\frac{dr_{\text{void}}}{dt}
= \frac{a^2}{z_c\, r_{\text{void}}}
\Big[
\omega_v C_v
- \omega_i C_i
- \omega_v\, e_v\!\left(\exp\!\left(\frac{2\gamma \Omega}{r_{\text{void}} k_B T}\right)-1\right)
\Big].
$$


$$
\frac{dC_{iL,i}^{111}}{dt}
= \frac{2\pi a^{2}}{z_c \Omega}\; r_{111}\, C_{iL}^{111}
(Z_{i}^{111}\omega_i (C_i + 2C_{2i} + 3C_{3i}) - Z_{v}^{111}\omega_v C_v)
$$

$$
\frac{dC_{iL,i}^{100}}{dt}
= \frac{2\pi a^{2}}{z_c \Omega}\; r_{100}\, C_{iL}^{100}
(Z_{i}^{100}\omega_i (C_i + 2C_{2i} + 3C_{3i}) - Z_{v}^{100}\omega_v C_v)
$$

$$
r_{111} = \sqrt{\frac{\Omega}{\pi b_{111}} \frac{C_{iL,i}^{111}}{C_{iL}^{111}}},
\quad b_{111} = \frac{\sqrt{3}}{2} a
$$

$$
r_{100} = \sqrt{\frac{\Omega}{\pi b_{100}} \frac{C_{iL,i}^{100}}{C_{iL}^{100}}},
\quad b_{100} = \frac{1}{2} a
$$


Collision cascade efficiency to generate a void:
$$\frac{dC_{void}}{dt} = \frac{\varepsilon_{void} G \Omega}{V_{void}} \left( 1 - \frac{V_{cap}}{\Omega} C_{void} \right) - \frac{C_{void}}{\tau_{void}}$$

to
$$
\frac{dC_{\text{void}}}{dt}
= G_{\text{void}}
\left( \frac{C_{\text{void}}}{C^{\text{void}}_{v}} \right)
\left\{ 1 - n_{\text{cap}}\, C_{\text{void}} \right\}
- \frac{C_{\text{void}}}{\tau_{\text{void}}}
$$
with $G_{\text{void}} = \varepsilon_{\text{void}}\,G, n_{\text{cap}} = \frac{V_{\text{cap}}}{\Omega}.$

### Fusion Modified (He gas)

$$
\frac{dC_b}{dt} = k_{\text{nuc}}^{(b)}(T)\, C_v C_g - \frac{C_b}{\tau_{\text{diss}}^{(b)}(T)}
$$

- $C_b$: bubble number density  
- $C_g$: helium concentration  
- $k_{\text{nuc}}^{(b)}(T)$: helium-assisted nucleation rate  
- $\tau_{\text{diss}}^{(b)}(T)$: bubble dissolution time

$$
\frac{dm_b}{dt} = \omega_g C_g - \omega_{\text{irr}}\, m_b
$$

- $m_b$: average number of He atoms per bubble
- $\omega_g$: helium capture frequency  
- $\omega_{\text{irr}}$: irradiation re-solution rate


$$
\frac{dR_b}{dt} = \frac{a^2}{z_c R_b} [\omega_v C_v - \omega_i C_i - \omega_v\,(e_{v,b} - e_v)]
$$

$$
e_{v,b} = e_v \exp(\frac{(P_g - \frac{2\gamma}{R_b})\,\Omega}{k_B T}),  P_g = \frac{m_b k_B T}{\frac{4}{3}\pi R_b^3}
$$

- $R_b$: bubble radius  

$$
\frac{dC_g}{dt} = G_{\text{He}} - \omega_g C_g C_b - k_{\text{nuc}}^{(b)}(T)\ C_v C_g + \omega_{\text{irr}}\, m_b C_b - k_{gS} C_g
$$

- $C_g$: helium concentration
- $G_{\text{He}}$: helium generation rate
- If bubbles absorb He → Cg decreases, m_b increases
- Creation of new bubbles removes He
- If cascades re-solve He → Cg increases, m_b decreases
- He lost to grain boundaries, sinks, dislocations



### Conversion of the ⟨111⟩ and ⟨100⟩ loop equations to the compact form

Compare Eq. (3) with Eq. (C): both have the same dimensional structure.  
Hence we can define

$$
\frac{\ell'_{hkl}}{\ell}
= \frac{2a^{2}}{z_c}
\sqrt{\frac{\pi}{b_{hkl}\Omega}}
$$

so that each type of loop has its own effective ratio:

$$
\frac{\ell'_{111}}{\ell}, \qquad
\frac{\ell'_{100}}{\ell}.
$$

Then Eq. (1) can be written in the same compact form as Eq. (92):

$$
\frac{dC^{hkl}_{i,iL}}{dt}
=
\left( \frac{\ell'_{hkl}}{\ell} \right)
\sqrt{ C^{hkl}_{i,iL} \, C^{hkl}_{a\,iL} }
\left\{
Z^{hkl}_i (\phi_i + \phi_{2i} + \phi_{3i})
- Z^{hkl}_v \phi_v
\right\}.
$$


### Vacancy Emission (Release) Term

$$
-\omega_v [ C_v^{eq}(R) - C_v^{eq}(\infty) ]
$$

is the **vacancy emission (or release) term** — it represents how many vacancies are **thermally emitted** from the void surface because the local equilibrium concentration at the curved interface is higher than in the flat bulk.

---

### Equilibrium Concentrations

- $C_v^{eq}(\infty)$: equilibrium vacancy concentration in a flat (unpressurized) bulk.  
- $C_v^{eq}(R)$: equilibrium vacancy concentration right at the curved void surface.  

It’s larger for small voids because of **Gibbs–Thomson curvature** and/or internal gas pressure:

$$
C_v^{eq}(R) = C_v^{eq}(\infty)
\exp\left(\frac{2\gamma\Omega}{RkT} - \frac{p_g\Omega}{kT}\right)
$$

The difference $C_v^{eq}(R) - C_v^{eq}(\infty)$ therefore measures the **vacancy supersaturation** at the surface created by surface tension (and pressure).

---

### 💡 For a Void (No Gas)

If $p_g = 0$, emission arises only from **curvature**:
- Small voids (large curvature) have higher $C_v^{eq}(R)$, thus emit vacancies faster and tend to **shrink**.  
- At large $R$, curvature becomes negligible ($C_v^{eq}(R) \approx C_v^{eq}(\infty)$), and emission **vanishes**.
