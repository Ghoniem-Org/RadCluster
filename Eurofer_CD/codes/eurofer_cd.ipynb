# Eurofer_CD — Cluster Dynamics for bcc Fe / EUROFER97

Mean-field cluster-size–resolved rate equations for radiation-induced defect cluster evolution in EUROFER97 / ferritic-martensitic steel, with explicit He–vacancy kinetics.

**Physics**: Ghoniem, N.M. (2024), *Formulation of Cluster Dynamics Equations for Irradiated Ferritic-Martensitic Steels* — Sections 1–5.

**He-vacancy modes** (PDF Section 5.6.5):
- `decoupled`: He shifts void binding energy only (fission, ~0.5 appm He/dpa)
- `fast_eq`: He mass balance per void class (general / fusion, ~10 appm He/dpa)

## 1 — Python LSODA solver: 'decoupled' He mode (fission)

```python
import sys
sys.path.insert(0, '../../')

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from Eurofer_CD.py_utils.simulation    import EuroferCDSimulation
from Eurofer_CD.py_utils.visualization import EuroferCDVisualizer, create_run_directory, write_provenance

# Configuration — override Excel values if desired
SIM_CONFIG = {
    'Ni':      100,
    'Nv':      100,
    'he_mode': 'decoupled',   # 'decoupled' | 'fast_eq'
}

sim     = EuroferCDSimulation(**SIM_CONFIG)
results = sim.run_simulation(
    t_span     = (1e-8, 1e7),
    n_segments = 60,
    rtol       = 1e-4,
    atol       = 1e-16,
    verbose    = True,
)

# Save results and plots
run_dir = create_run_directory(label='py_decoupled')
write_provenance(run_dir, results, sim.input_data)

viz = EuroferCDVisualizer(results, sim.input_data, run_dir=run_dir)
viz.plot_all()
print(f'Output: {run_dir}')

t     = results['time']
c     = results['concentrations']
tot   = results['totals']
sw    = results['swelling']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Point defects
ax = axes[0, 0]
ax.loglog(t, np.maximum(c['Ci1'], 1e-40), label='$C_{i1}$')
ax.loglog(t, np.maximum(c['Cv1'], 1e-40), label='$C_{v1}$')
ax.loglog(t, np.maximum(c['C_He'], 1e-40), label='$C_{He}$', ls='--')
ax.set_xlabel('Time (s)'); ax.set_ylabel('Concentration'); ax.set_title('Point defects')
ax.set_ylim(bottom=1e-18)
ax.legend(); ax.grid(True, which='both', alpha=0.3)

# Total defect content
ax = axes[0, 1]
ax.loglog(t, np.maximum(tot['total_i'], 1e-40), label='Total SIA')
ax.loglog(t, np.maximum(tot['total_v'], 1e-40), label='Total vacancy', ls='--')
ax.set_xlabel('Time (s)'); ax.set_ylabel('Total content'); ax.set_title('Total defect inventories')
ax.set_ylim(bottom=1e-18)
ax.legend(); ax.grid(True, which='both', alpha=0.3)

# Mean cluster sizes
ax = axes[1, 0]
ms = results['mean_sizes']
ax.semilogx(t, ms['mean_n'], label='⟨n⟩ SIA (clusters, n≥2)')
ax.semilogx(t, ms['mean_m'], label='⟨m⟩ vac (clusters, m≥2)', ls='--')
ax.set_xlabel('Time (s)'); ax.set_ylabel('Mean size'); ax.set_title('Mean cluster sizes')
ax.legend(); ax.grid(True, which='both', alpha=0.3)

# Void swelling
ax = axes[1, 1]
ax.semilogx(t, sw * 100, color='C3')
ax.set_xlabel('Time (s)'); ax.set_ylabel('Swelling (%)'); ax.set_title('Void swelling')
ax.grid(True, which='both', alpha=0.3)

plt.tight_layout()
plt.show()

```

## 3 — 'fast_eq' He mode (for fusion or higher He/dpa)

```python
sim_feq     = EuroferCDSimulation(Ni=100, Nv=100, he_mode='fast_eq')
results_feq = sim_feq.run_simulation(t_span=(1e-8, 1e6), n_segments=60, verbose=True)
```

```python
# Compare free He and swelling between modes
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

ax = axes[0]
ax.loglog(t, np.maximum(results['totals']['C_He'], 1e-40), label='decoupled')
ax.loglog(results_feq['time'], np.maximum(results_feq['totals']['C_He'], 1e-40),
          label='fast_eq', ls='--')
ax.set_xlabel('Time (s)'); ax.set_ylabel('Free He (atom frac)')
ax.set_title('Free He: mode comparison')
ax.set_ylim(bottom=1e-18)
ax.legend(); ax.grid(True, alpha=0.3)

ax = axes[1]
ax.semilogx(t, results['swelling'] * 100, label='decoupled')
ax.semilogx(results_feq['time'], results_feq['swelling'] * 100, label='fast_eq', ls='--')
ax.set_xlabel('Time (s)'); ax.set_ylabel('Swelling (%)')
ax.set_title('Void swelling: mode comparison'); ax.legend(); ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4 — Binding energy and rate constant inspection

```python
from Eurofer_CD.py_utils.binding_energies import E_b_void, E_b_loop, E_b_He

ms = np.arange(1, 101)
ns = np.arange(2, 101)

inp = sim.input_data
E_f_v   = float(inp.physical_props['E_f_v'])
gamma_s = float(inp.physical_props['gamma_s'])
Omega   = float(inp.physical_props['Omega'])
E_b_2i  = float(inp.physical_props['E_b_2i'])
E_b_inf = float(inp.physical_props['E_b_inf_loop'])

Eb_void_arr = np.array([E_b_void(m, E_f_v, gamma_s, Omega) for m in ms])
Eb_loop_arr = np.array([E_b_loop(n, E_b_2i, E_b_inf) for n in ns])

fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

ax = axes[0]
ax.semilogx(ms, Eb_void_arr)
ax.axhline(E_f_v, ls='--', color='gray', label=f'$E_f^v$ = {E_f_v} eV')
ax.set_xlabel('Vacancy cluster size m'); ax.set_ylabel('Binding energy (eV)')
ax.set_title('Void binding energy $E_b^{void}(m)$')
ax.legend(); ax.grid(True, alpha=0.3)

ax = axes[1]
ax.semilogx(ns, Eb_loop_arr)
ax.axhline(E_b_inf, ls='--', color='gray', label=f'$E_b^{{\\infty}}$ = {E_b_inf} eV')
ax.set_xlabel('SIA cluster size n'); ax.set_ylabel('Binding energy (eV)')
ax.set_title('Interstitial loop binding $E_b^{loop}(n)$')
ax.legend(); ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
