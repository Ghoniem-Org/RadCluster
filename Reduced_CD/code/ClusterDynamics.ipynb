# ClusterDynamics — Ferritic-Martensitic Steel

Cluster-size resolved rate equations for radiation-induced defect clustering in **EUROFER97 / ferritic-martensitic steel**.

## Physics summary

Tracks the full number-density distribution of vacancy clusters $f_v(n,t)$ and interstitial clusters $f_i(n,t)$ up to maximum sizes $N_v$ and $N_i$, together with:
- 1/2⟨111⟩ and ⟨100⟩ interstitial loop densities and mean radii
- Void number density and mean radius

The size-space master equations (Ghoniem & Cho 1974) are:

$$
\frac{df_v(n)}{dt} = J_v(n-1\to n) - J_v(n\to n+1) + S_v(n)
$$

$$
J_v(n\to n+1) = \beta_v(n)\,f_v(n) - \alpha_v(n+1)\,f_v(n+1)
$$

with absorption rate $\beta_v(n) = 4\pi r_n D_v / \Omega$ and emission rate $\alpha_v(n) = \beta_v(n-1)\exp(-E_b(n)/k_BT)$.

Loop and void evolution are inherited from the `EuroferMicrostructure` mean-field model (see `../CLAUDE.md`).

## Code structure

```
ClusterDynamics/py_utils/
  input_data.py      — InputData (3-sheet Excel loader + derived params)

  simulation.py      — ClusterDynamicsSimulation + run_cluster_dynamics_simulation
  post_process.py    — mean size, swelling, summary_table

```

````python
%load_ext autoreload
%autoreload 2

import sys
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
from pathlib import Path

sys.path.insert(0, '../')   # make ClusterDynamics/ importable

from py_utils.simulation import ClusterDynamicsSimulation
from py_utils.post_process import summary_table, size_distribution_at_dose
from py_utils.visualization import plot_all

# ── Configuration ──────────────────────────────────────────────────────────────
TAGS = ['ion', '200C']   # output directory tags — edit before running

SIMULATION_CONFIG = {
    't_begin':  1e-1,    # Start time [s]
    't_end':    1e6,     # End time [s]  (≈ 10 dpa at 1e-5 dpa/s neutron rate)
    'n_points': 500,     # Output time points
    'method':   'LSODA', # ODE solver
    'rtol':     1e-6,
    'atol':     1e-20,
    'log_time': True,
}

# ── Output directory ───────────────────────────────────────────────────────────
timestamp  = datetime.now().strftime('%Y%m%d_%H%M%S')
tag_str    = '_'.join(TAGS) if TAGS else 'run'
output_dir = Path('../output') / f'{timestamp}_{tag_str}'
output_dir.mkdir(parents=True, exist_ok=True)
print(f"Output directory: {output_dir.resolve()}")

# ── Run simulation ─────────────────────────────────────────────────────────────
results, sim = None, None
try:
    sim     = ClusterDynamicsSimulation()
    results = sim.run(
        t_begin  = SIMULATION_CONFIG['t_begin'],
        t_end    = SIMULATION_CONFIG['t_end'],
        n_points = SIMULATION_CONFIG['n_points'],
        method   = SIMULATION_CONFIG['method'],
        rtol     = SIMULATION_CONFIG['rtol'],
        atol     = SIMULATION_CONFIG['atol'],
        log_time = SIMULATION_CONFIG['log_time'],
    )
    sim.save(results, output_dir)
except Exception as e:
    import traceback
    print(f"Fatal error: {e}")
    traceback.print_exc()

# ── Plots ──────────────────────────────────────────────────────────────────────
if results is not None:
    plot_all(
        results,
        sim.input_data,
        output_dir=output_dir / 'plots',
        x_axis='dpa',
    )

# ── Summary table ──────────────────────────────────────────────────────────────
if results is not None:
    Omega = sim.input_data.physical_props['Omega']
    df    = summary_table(results, Omega)
    display(df.tail(10))

# ── Cluster size distributions at selected doses ───────────────────────────────
if results is not None:
    target_dpas = [0.01, 0.1, 1.0, 5.0]
    fig, axes   = plt.subplots(1, 2, figsize=(12, 4))
    colors      = plt.cm.viridis(np.linspace(0.2, 0.9, len(target_dpas)))

    for dpa_t, col in zip(target_dpas, colors):
        fv_s, fi_s, dpa_a = size_distribution_at_dose(results, dpa_t)
        n_v = sorted(fv_s.keys())
        n_i = sorted(fi_s.keys())
        axes[0].semilogy(n_v, [fv_s[n] for n in n_v], 'o-',  color=col, label=f'{dpa_a:.2f} dpa')
        axes[1].semilogy(n_i, [fi_s[n] for n in n_i], 's--', color=col, label=f'{dpa_a:.2f} dpa')

    axes[0].set_xlabel('Cluster size n');  axes[0].set_ylabel('$f_v(n)$ (at/at)')
    axes[0].set_title('Vacancy cluster distribution');     axes[0].legend()
    axes[1].set_xlabel('Cluster size n');  axes[1].set_ylabel('$f_i(n)$ (at/at)')
    axes[1].set_title('Interstitial cluster distribution'); axes[1].legend()
    plt.tight_layout()

    dist_file = output_dir / 'plots' / 'cluster_distributions.png'
    dist_file.parent.mkdir(parents=True, exist_ok=True)
    plt.savefig(dist_file, dpi=150, bbox_inches='tight')
    plt.show()
    print(f"Distribution plot saved → {dist_file}")

# ── Parametric study (temperature sweep) — uncomment to run ───────────────────
# from py_utils.reaction_rates import ClusterRates
# from py_utils.rate_equations import ClusterDynamicsODE
#
# temperatures = [523, 573, 623, 673]  # K
# all_results  = {}
#
# for T in temperatures:
#     sim_T = ClusterDynamicsSimulation()
#     sim_T.input_data.material_params['T'] = T
#     sim_T.input_data.calculate_derived_parameters()
#     sim_T.rates = ClusterRates(sim_T.input_data)
#     sim_T.ode   = ClusterDynamicsODE(sim_T.input_data, sim_T.rates)
#     r = sim_T.run(**{k: SIMULATION_CONFIG[k] for k in
#                     ['t_begin','t_end','n_points','method','rtol','atol','log_time']})
#     all_results[T] = r
#     sim_T.save(r, output_dir / f'T{T}K')
#     print(f"T={T} K done.")
````
