# Radiation-Induced Microstructural Features in Eurofer




#### 1. Dislocation Loops
- Density: ~1e22–1e23 m^-3
- Size: 5–50 nm
- Mostly 1/2<111> and <100> loops. Saturate at ~1e23 m^-3, recover >450 °C.

#### 2. Voids (without He)
- Density: ~1e19–1e20 m^-3
- Size: 2–10 nm
- Very rare. Swelling <0.5% even at 100–150 dpa.

#### 3. Helium Bubbles
- **Matrix:** 1e22–1e23 m^-3, size 1–3 nm. Form at 250–450 °C, 10–30 dpa.
- **Grain boundaries / carbides:** 1e21–1e22 m^-3, size 5–20 nm.

#### 4. Precipitates
- **MX (V,Ta carbonitrides):** Initially 5–20 nm at ~1e23–1e24 m^-3. Dissolve or shrink <5 nm with dose >30 dpa.
- **M23C6 (Cr-rich carbides):** 50–150 nm initially, coarsen to 100–300 nm. Density ~1e20–1e21 m^-3.

#### 5. Summary
- Loops dominate defect structure at 250–400 °C, hardening material.
- Voids rare → Eurofer resists swelling.
- He bubbles drive embrittlement and intergranular fracture risk.
- Precipitate redistribution and coarsening degrade strength at high dose.

![Schematic of feature density vs size](../Materials/Eurofer/Figures/eurofer_microstructure_schematic.png)


| Feature                  | Density (m⁻³)        | Size (nm)   | Notes                                        |
|---------------------------|----------------------|-------------|----------------------------------------------|
| Dislocation loops         | 10²² – 10²³         | 5 – 50      | Saturate ~10²³, anneal >450 °C               |
| Voids (no He)             | 10¹⁹ – 10²⁰         | 2 – 10      | Rare, swelling <0.5% up to 150 dpa           |
| He bubbles (matrix)       | 10²² – 10²³         | 1 – 3       | Form with 10–30 dpa, 250–450 °C              |
| He bubbles (GB/carbides)  | 10²¹ – 10²²         | 5 – 20      | Grain boundary/interface decoration           |
| MX precipitates           | 10²³ – 10²⁴ → lower | 5–20 → <5   | Dissolution at >30 dpa                       |
| M₂₃C₆ carbides            | 10²⁰ – 10²¹         | 50 – 300    | Coarsen at 300–500 °C                        |

```python
# =====================================================================
# Radiation microstructure: loop & void size/density vs dose
#
# Reads Ferritics_Irradiated_Microstructure_Data.xlsx and produces
# 12 SEPARATE plots (size & density x loops & voids x neutron / ion /
# combined). Each plot:
#   - flags clearly outlying points and labels them "outlier"
#   - excludes outliers from the trend-line fit and +/-1 sigma envelope
#   - lists the source references (last name, year, abbreviated journal)
#     contributing to the plotted data, so each point is verifiable.
#   - legend entries combine the temperature bin with the source
#     (e.g. "250-300C Reith 2019 JNM").
# Outlier rule: in log10(dose)-log10(value) space, points whose residual
# from an OLS linear fit exceeds 2.0 standard deviations are flagged.
# All figures are saved under EuroferGraphics/microstructure/.
# =====================================================================
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

XLSX = Path('..') / '..' / 'input' / 'Ferritics_Irradiated_Microstructure_Data.xlsx'
OUT  = Path('..') / '..' / 'output' / 'microstructure'
OUT.mkdir(parents=True, exist_ok=True)

FONT_SIZE = 16

# Map of full journal names -> short abbreviations to keep references compact.
JOURNAL_ABBREV = [
    ('journal of nuclear materials',          'JNM'),
    ('nuclear materials and energy',          'NME'),
    ('acta materialia',                       'Acta Mater.'),
    ('scripta materialia',                    'Scripta Mater.'),
    ('philosophical magazine',                'Phil. Mag.'),
    ('materials science and engineering',     'MSE'),
    ('nuclear instruments and methods',       'NIM'),
    ('fusion engineering and design',         'FED'),
    ('journal of materials research',         'JMR'),
]

def abbrev_journal(s):
    if not isinstance(s, str): return ''
    low = s.lower()
    for full, short in JOURNAL_ABBREV:
        if full in low:
            return short
    # Fallback: take the first three significant words, capitalised
    words = [w for w in re.split(r'\s+', s.strip()) if w and w.lower() not in {'of','and','the','for','on','in'}]
    return ' '.join(words[:2])

def load_journal_map(xlsx_path):
    """Build {first-author-year -> journal-abbrev} from the References sheet."""
    try:
        ref_df = pd.read_excel(xlsx_path, sheet_name='References', header=1)
    except Exception:
        return {}
    mapping = {}
    cols = {c.lower().strip(): c for c in ref_df.columns if isinstance(c, str)}
    author_c = next((cols[k] for k in cols if 'author' in k), None)
    year_c   = next((cols[k] for k in cols if 'year'   in k), None)
    journ_c  = next((cols[k] for k in cols if 'journal' in k or 'source' in k), None)
    if not (author_c and year_c and journ_c):
        return {}
    for _, r in ref_df.iterrows():
        a = r.get(author_c); y = r.get(year_c); j = r.get(journ_c)
        if pd.isna(a) or pd.isna(y): continue
        first = str(a).split(',')[0].split(' and ')[0].split('&')[0].strip()
        first = re.sub(r'\s+et al\.?$', '', first, flags=re.I).strip()
        try:    yr = str(int(float(y)))
        except (ValueError, TypeError): yr = str(y).strip()
        mapping[f'{first} {yr}'] = abbrev_journal(str(j))
    return mapping

JOURNAL_MAP = load_journal_map(XLSX)

def short_ref(raw_ref):
    """'Klimenkov et al. 2012' -> 'Klimenkov 2012 JNM' when journal known."""
    if not isinstance(raw_ref, str) or not raw_ref.strip():
        return ''
    s = raw_ref.strip()
    s = re.sub(r'\s+et al\.?', '', s, flags=re.I)
    s = re.sub(r'\s*&\s*\S+', '', s)
    s = re.sub(r'\s+', ' ', s).strip()
    key = s
    jrnl = JOURNAL_MAP.get(key, '')
    return f'{s} {jrnl}'.strip()

def parse_numeric(x):
    """Return mean numeric value from a possibly-fuzzy cell ('~7-15', '<5', '12.9 Å')."""
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return np.nan
    if isinstance(x, (int, float)):
        return float(x)
    s = str(x).strip().replace('×10','e').replace('·10','e')
    s = s.replace('—','-').replace('–','-').replace('−','-')
    s = re.sub(r'[<>~≈≤≥]', '', s)
    s = re.sub(r'(?<=\d)\s*-\s*(?=\d)', ' ', s)
    nums = re.findall(r'\d+\.?\d*(?:[eE][-+]?\d+)?', s)
    if not nums:
        return np.nan
    vals = [float(n) for n in nums[:2]]
    return float(np.mean(vals))

TEMP_BINS = [
    ('<200°C',     -np.inf, 200,  'o', '#1f77b4'),
    ('200-250°C',  200, 250,      's', '#17becf'),
    ('250-300°C',  250, 300,      'D', '#2ca02c'),
    ('300-350°C',  300, 350,      '^', '#bcbd22'),
    ('350-400°C',  350, 400,      'v', '#ff7f0e'),
    ('400-450°C',  400, 450,      'P', '#d62728'),
    ('450-500°C',  450, 500,      'X', '#9467bd'),
]
def bin_label(T):
    if pd.isna(T): return None
    for lab, lo, hi, *_ in TEMP_BINS:
        if lo <= T < hi: return lab
    return None

def load_sheet(name):
    df = pd.read_excel(XLSX, sheet_name=name, header=[1,2])
    df.columns = [str(c2) for (_, c2) in df.columns]
    df = df.dropna(how='all').reset_index(drop=True)
    return df

def find_col(df, *needles):
    for c in df.columns:
        s = str(c).lower().replace('\n',' ')
        if all(n.lower() in s for n in needles): return c
    return None

def build_records(df, kind):
    dose_c = find_col(df, 'dose')
    temp_c = find_col(df, 'temp')
    ref_c  = find_col(df, 'reference')
    if   kind == 'loop_size':    val_c = find_col(df, 'loop',   'diam')
    elif kind == 'loop_density': val_c = find_col(df, 'loop',   'density')
    elif kind == 'void_size':    val_c = find_col(df, 'cavity', 'diam')
    elif kind == 'void_density': val_c = find_col(df, 'cavity', 'density')
    else: raise ValueError(kind)
    if val_c is None or dose_c is None or temp_c is None:
        return pd.DataFrame(columns=['dose','temp','val','bin','ref'])
    rows = []
    for _, r in df.iterrows():
        d = parse_numeric(r[dose_c])
        t = parse_numeric(r[temp_c])
        v = parse_numeric(r[val_c])
        if np.isnan(d) or np.isnan(v) or np.isnan(t): continue
        b = bin_label(t)
        if b is None: continue
        ref_raw = r[ref_c] if ref_c is not None else ''
        rows.append((d, t, v, b, short_ref(ref_raw)))
    return pd.DataFrame(rows, columns=['dose','temp','val','bin','ref'])

def detect_outliers(x, y, z_thresh=2.0):
    """Flag outliers from an OLS log-log fit. Returns boolean mask (True = outlier)."""
    mask = (x > 0) & (y > 0) & np.isfinite(x) & np.isfinite(y)
    out = np.zeros(len(x), dtype=bool)
    if mask.sum() < 4:
        return out
    lx = np.log10(x[mask]); ly = np.log10(y[mask])
    slope, intercept = np.polyfit(lx, ly, 1)
    residuals = ly - (slope*lx + intercept)
    sd = np.std(residuals, ddof=1)
    if sd == 0 or not np.isfinite(sd):
        return out
    out[np.where(mask)[0]] = np.abs(residuals)/sd > z_thresh
    return out

def save_legend_figure(handles, labels, title, data_fname):
    """Render a separate figure containing only the legend for the data plot."""
    if not handles:
        return
    n = len(labels)
    ncol = 1 if n <= 12 else 2 if n <= 24 else 3
    nrows = -(-n // ncol)
    fig_h = max(2.0, 0.45 * nrows + 1.2)
    fig_w = 6.0 * ncol
    fig = plt.figure(figsize=(fig_w, fig_h))
    ax = fig.add_subplot(111)
    ax.axis('off')
    ax.set_title(f'{title} - legend', fontsize=FONT_SIZE)
    leg = ax.legend(handles, labels, loc='center', ncol=ncol,
                    fontsize=FONT_SIZE, frameon=True, framealpha=0.95,
                    handlelength=2.0, handletextpad=0.6, borderpad=0.8)
    leg.get_frame().set_edgecolor('#888888')
    stem = data_fname.rsplit('.', 1)[0]
    ext = data_fname.rsplit('.', 1)[1] if '.' in data_fname else 'png'
    legend_fname = f'{stem}_Legend.{ext}'
    fig.savefig(OUT/legend_fname, dpi=150, bbox_inches='tight')
    plt.show(); plt.close(fig)

def plot_single(df, title, ylabel, fname, ylog=False):
    fig, ax = plt.subplots(figsize=(9, 6))
    if df.empty:
        ax.set_title(f'{title}\n(no data)', fontsize=FONT_SIZE)
        ax.set_xlabel('Dose (dpa)', fontsize=FONT_SIZE)
        ax.set_ylabel(ylabel, fontsize=FONT_SIZE)
        ax.tick_params(axis='both', labelsize=FONT_SIZE)
        plt.tight_layout()
        fig.savefig(OUT/fname, dpi=150, bbox_inches='tight')
        plt.show(); plt.close(fig)
        save_legend_figure([], [], title, fname)
        return

    x = df['dose'].values.astype(float)
    y = df['val'].values.astype(float)
    outlier = detect_outliers(x, y, z_thresh=2.0)
    keep = ~outlier

    # Inliers grouped by temperature bin and source (legend lists each pair)
    for lab, lo, hi, marker, color in TEMP_BINS:
        sub = df[(df['bin']==lab) & keep]
        if sub.empty: continue
        for ref, grp in sub.groupby('ref', sort=True):
            legend_label = f'{lab} {ref}' if isinstance(ref, str) and ref.strip() else lab
            ax.scatter(grp['dose'], grp['val'], s=110, marker=marker,
                       facecolors=color, edgecolors='black', linewidths=0.7,
                       alpha=0.9, label=legend_label, zorder=3)

    # Outliers: red X with explicit "outlier" label
    if outlier.any():
        ax.scatter(x[outlier], y[outlier], s=180, marker='x',
                   color='red', linewidths=2.2, zorder=4,
                   label=f'outlier (n={int(outlier.sum())})')
        for xi, yi in zip(x[outlier], y[outlier]):
            ax.annotate('outlier', xy=(xi, yi),
                        xytext=(6, 6), textcoords='offset points',
                        fontsize=FONT_SIZE, color='red', fontweight='bold')

    # Trend line + +/-1 sigma envelope from inliers only
    xk = x[keep]; yk = y[keep]
    fm = (xk>0) & (yk>0) & np.isfinite(xk) & np.isfinite(yk)
    xk = xk[fm]; yk = yk[fm]
    if len(xk) >= 2 and np.ptp(xk) > 0:
        lx = np.log10(xk); ly = np.log10(yk)
        slope, intercept = np.polyfit(lx, ly, 1)
        residuals = ly - (slope*lx + intercept)
        sd = float(np.std(residuals, ddof=1)) if len(residuals) > 1 else 0.0
        xs = np.logspace(np.log10(xk.min()), np.log10(xk.max()), 80)
        ly_mid = slope*np.log10(xs) + intercept
        y_mid = 10**ly_mid
        y_hi  = 10**(ly_mid + sd)
        y_lo  = 10**(ly_mid - sd)
        ax.plot(xs, y_mid, '-',  color='black', lw=1.6,
                label=f'trend (log-log fit, n={len(xk)})')
        ax.plot(xs, y_hi,  '--', color='gray',  lw=1.0, label='+1σ')
        ax.plot(xs, y_lo,  ':',  color='gray',  lw=1.0, label='-1σ')
        ax.fill_between(xs, y_lo, y_hi, color='gray', alpha=0.12, zorder=1)

    if not ylog:
        ymin, _ = ax.get_ylim()
        if ymin < 0:
            ax.set_ylim(bottom=0)
    ax.set_xlabel('Dose (dpa)', fontsize=FONT_SIZE)
    ax.set_ylabel(ylabel, fontsize=FONT_SIZE)
    ax.set_title(title, fontsize=FONT_SIZE)
    ax.tick_params(axis='both', labelsize=FONT_SIZE)
    ax.set_xscale('log')
    if ylog: ax.set_yscale('log')
    ax.grid(True, which='both', ls=':', alpha=0.4)
    # Capture legend entries for the companion figure, but do NOT draw a legend here.
    handles, labels = ax.get_legend_handles_labels()
    fig.savefig(OUT/fname, dpi=150, bbox_inches='tight')
    plt.show(); plt.close(fig)

    # Companion figure: legend only.
    save_legend_figure(handles, labels, title, fname)

# ---- Load sheets and build datasets ---------------------------------
n_loops = load_sheet('N – Dislocation Loops')
n_voids = load_sheet('N – Voids & Swelling')
i_loops = load_sheet('Ion – Dislocation Loops')
i_voids = load_sheet('Ion – Voids & Swelling')

data = {
    ('N',  'loop_size'):    build_records(n_loops, 'loop_size'),
    ('N',  'loop_density'): build_records(n_loops, 'loop_density'),
    ('N',  'void_size'):    build_records(n_voids, 'void_size'),
    ('N',  'void_density'): build_records(n_voids, 'void_density'),
    ('Ion','loop_size'):    build_records(i_loops, 'loop_size'),
    ('Ion','loop_density'): build_records(i_loops, 'loop_density'),
    ('Ion','void_size'):    build_records(i_voids, 'void_size'),
    ('Ion','void_density'): build_records(i_voids, 'void_density'),
}
for k, df in data.items():
    print(f'{k}: {len(df)} usable rows')

def combined(kind):
    return pd.concat(
        [data[('N',kind)].assign(src='Neutron'),
         data[('Ion',kind)].assign(src='Ion')],
        ignore_index=True,
    )

# ---- Twelve separate plots ------------------------------------------
plots = [
    (data[('N',  'loop_size')],    'Neutron — Dislocation loop size vs dose',
     'Mean loop diameter (nm)',         'Neutron_Loop_Size.png',     False),
    (data[('N',  'loop_density')], 'Neutron — Dislocation loop density vs dose',
     'Loop density (×10²² m⁻³)',        'Neutron_Loop_Density.png',  True),
    (data[('N',  'void_size')],    'Neutron — Void/cavity size vs dose',
     'Mean cavity diameter (nm)',       'Neutron_Void_Size.png',     False),
    (data[('N',  'void_density')], 'Neutron — Void/cavity density vs dose',
     'Cavity density (×10²² m⁻³)',      'Neutron_Void_Density.png',  True),
    (data[('Ion','loop_size')],    'Ion — Dislocation loop size vs dose',
     'Mean loop diameter (nm)',         'Ion_Loop_Size.png',         False),
    (data[('Ion','loop_density')], 'Ion — Dislocation loop density vs dose',
     'Loop density (×10²² m⁻³)',        'Ion_Loop_Density.png',      True),
    (data[('Ion','void_size')],    'Ion — Void/cavity size vs dose',
     'Mean cavity diameter (nm)',       'Ion_Void_Size.png',         False),
    (data[('Ion','void_density')], 'Ion — Void/cavity density vs dose',
     'Cavity density (×10²² m⁻³)',      'Ion_Void_Density.png',      True),
    (combined('loop_size'),        'Combined (Neutron + Ion) — Loop size vs dose',
     'Mean loop diameter (nm)',         'Combined_Loop_Size.png',    False),
    (combined('loop_density'),     'Combined (Neutron + Ion) — Loop density vs dose',
     'Loop density (×10²² m⁻³)',        'Combined_Loop_Density.png', True),
    (combined('void_size'),        'Combined (Neutron + Ion) — Void size vs dose',
     'Mean cavity diameter (nm)',       'Combined_Void_Size.png',    False),
    (combined('void_density'),     'Combined (Neutron + Ion) — Void density vs dose',
     'Cavity density (×10²² m⁻³)',      'Combined_Void_Density.png', True),
]

for df, title, ylabel, fname, ylog in plots:
    plot_single(df, title, ylabel, fname, ylog=ylog)
    print(f'wrote {OUT/fname}')
```
